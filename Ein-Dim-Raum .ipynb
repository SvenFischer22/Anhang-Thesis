# 1-Dimensionales Problem

```python
# List von Produkten 

products = ['a0', 'a1', 'a2', 'a3', 'a4', 'a5', 'a6', 'a7', 'alang', 'q5', 'q6']

# Fläche der Produkte die wir von der geammt möglichen fläche abzihen können 

profit = {'a0': 999949, 
         'a1': 499554,
         'a2': 249480,
         'a3': 124740,
         'a4': 62370,
         'a5': 31080,
         'a6': 15540,
         'a7': 7770,
         'alang': 22050,
         'q5': 21904,
         'q6': 10716}

# Fläche welche durch die Objekte verbraucht wird

consumption = {'space':
               {'a0': 999949, 
         'a1': 499554,
         'a2': 249480,
         'a3': 124740,
         'a4': 62370,
         'a5': 31080,
         'a6': 15540,
         'a7': 7770,
         'alang': 22050,
         'q5': 21904,
         'q6': 10716}
              }

# Maximal mögliche nutzfläche 

resources = {
    'space': 2378000
}

# erstellung eines Maximierungsproblemes / oder alternativ minimierung dann aber Maximalfe fläche - nutzfläche 

prob = LpProblem("1-Dim", LpMaximize)

# Produkte als Variablen 

x = LpVariable.dicts("Product",products,0,1,cat="Integer")
obj = LpVariable("profit")

# Maximierung

prob += obj == lpSum([profit[i]*x[i] for i in products])


prob += obj, "Total Profit of Production"


# Nebenbedingung 1 Maxfläche >= genutzt fläche

for resource in resources.keys():
    prob += lpSum([x[i]*consumption[resource][i] for i in products]) <= resources[resource], "{}".format(resource)

prob += obj <= resources["space"]
print(prob)
# Optimize

prob.solve()

# Print the status of the solved LP
print("Status = %s" % LpStatus[prob.status])

# Print the value of the variables at the optimum
for i in products:
    print("%s = %f" % (x[i].name, x[i].varValue))

# Print the value of the objective
print("Objective = %f" % value(prob.objective))
```
